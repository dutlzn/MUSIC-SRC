```python
import librosa
import numpy as np
import matplotlib.pyplot as plt 
import os
import librosa.display
```

```python
plt.figure(figsize=(5,3))
songname = '../genres/blues/blues.00000.au'
y, sr = librosa.load(songname)
spectogram = librosa.feature.melspectrogram(
    y=y, sr=sr, n_mels=128, n_fft=2048,
)
spectogram = librosa.power_to_db(spectogram, ref=np.max)
librosa.display.specshow(spectogram, x_axis='time', y_axis='mel')
plt.colorbar()
plt.title('blues.00000.au')
plt.tight_layout()
plt.savefig('blues.00000.1.png')
```
