```python
def total_parameters(net):
    total = 0
    for name, p in net.named_parameters():
        params = p.numel() if p.requires_grad else 0
        print(name, params)
        if '5x5' in name:
            continue
        total += params
    return total
total_parameters(net)
```

```python
import torch
import torch.nn as nn
```

```python
def make_conv(in_channel, out_channel, kernel_size, padding=0, pool_size=2):
    return nn.Sequential(
        nn.Conv2d(in_channel, out_channel, kernel_size, padding=padding),
        nn.BatchNorm2d(out_channel),
        nn.ReLU(),
        nn.MaxPool2d(pool_size),
    )

yao_cnn = nn.Sequential(
    make_conv(1, 16, 3, 1, 2),
    make_conv(16, 32, 3, 1, 2),
    make_conv(32, 64, 3, 1, 2),
    make_conv(64, 128, 3, 1, 4),
    make_conv(128, 64, 3, 1, 4),
    nn.Linear(256, 10),
)

yao_rnn = nn.Sequential(
    nn.Linear())

total_parameters(yao_cnn)
```
