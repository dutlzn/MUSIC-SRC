```python
import pickle

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
with open('scores.pkl', 'rb') as f:
    raw_scores = pickle.load(f)
raw_scores
```

```python
train_scores = {}
test_scores = {}

for num_segment, scores_ in raw_scores.items():
    for feature_type, combinations in scores_.items():
        for combination_type, results in combinations.items():
            key = '{}-{}-{}'.format(num_segment, feature_type, combination_type)
            train_scores[key] = np.array(results['train'])
            test_scores[key] = np.array(results['test'])

df = pd.DataFrame(test_scores)
df
```
