```python
import pickle

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
with open('scores.pkl', 'rb') as f:
    raw_scores = pickle.load(f)
raw_scores
```

```python
train_scores = {}
test_scores = []

for feature_type, combinations in raw_scores[1].items():
    for combination_type, results in combinations.items():
        key = '{}-{}'.format(feature_type, combination_type)
        train_scores[key] = np.array(results['train'])
        for score in results['test']:
            test_scores.append((key, score))
print(len(set([x[0] for x in test_scores])))
df = pd.DataFrame(test_scores, columns=['key', 'acc'])
plt.figure(figsize=(8, 16))
sns.boxplot(x='key', data=df, orient='h')
```
