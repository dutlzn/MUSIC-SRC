```python
import os
import os.path as osp
import sys
sys.path.append(osp.abspath('..'))

import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm_notebook

import config
from datasets.gtzan import GTZAN_SPEC

%load_ext autoreload
%autoreload 2
```

```python
# Random seeds
np.random.seed(1234)
segments = -1

num_principal_features = 130
sigmaPC = 1e-5
```

```python
train_set = GTZAN_SPEC(phase='train', segments=segments)
test_set = GTZAN_SPEC(phase='test', segments=segments)

print('Train:', len(train_set))
print('Test:', len(test_set))
print('Shape:', train_set[0][0].shape)
```

```python
def concatenate_data(*datasets):
    X = []
    for dataset in datasets:
        for sample_idx, (x, _) in tqdm_notebook(enumerate(dataset), total=len(dataset)):
            for x_ in x:
                X.append(x_)
    X = np.concatenate(X, axis=1)
    return X

X = concatenate_data(train_set, test_set)
X.shape
```

```python
# PCA whiten
Xcov = X.dot(X.T) / X.shape[1]
num_features = Xcov.shape[0]
eigen_values, eigen_vectors = np.linalg.eigh(Xcov)

E = eigen_vectors[:, num_features:num_features-num_principal_features-1:-1]
S = np.diag(eigen_values[num_features:num_features-num_principal_features-1:-1])

Xpc = E.T.dot(X)
Xrec = E.dot(Xpc)

Ewhiten = np.diag(1. / np.sqrt((np.diag(S)+sigmaPC))).dot(E.T)
Eunwhiten = E.dot(np.diag(np.sqrt(np.diag(S)+sigmaPC)))

Xrec2 = Eunwhiten.dot(Ewhiten).dot(X)
```

```python
plt.figure(figsize=(16, 32))
plt.subplot(1, 3, 1)
plt.imshow(X[:, :301], cmap='hot')
plt.subplot(1, 3, 2)
plt.imshow(Xrec[:, :301], cmap='hot')
plt.subplot(1, 3, 3)
plt.imshow(Xrec2[:, :301], cmap='hot')
plt.show()
```

```python
# aggregate over all frames, calculate simple summary statistics
# such as average, max or standard deviation for each channel
x_train = np.empty((750 * segments, num_principal_features * 3))
y_train = np.empty((750 * segments,))
x_test = np.empty((250 * segments, num_principal_features * 3))
y_test = np.empty((250 * segments,))

def populate_features(src_dataset, dst_x, dst_y):
    for sample_idx, (x, y) in tqdm_notebook(enumerate(src_dataset), total=len(src_dataset)):
        x = np.array([Ewhiten.dot(x_) for x_ in x])
        mean = np.mean(x, axis=2)
        max = np.max(x, axis=2)
        std = np.std(x, axis=2)
        start = sample_idx*segments
        end = (sample_idx + 1)*segments
        dst_x[start:end] = np.hstack((mean, max, std))
        for segment_idx in range(segments):
            dst_y[sample_idx*segments + segment_idx] = y
            
populate_features(train_set, x_train, y_train)
populate_features(test_set, x_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

params = {
    'C': [10**x for x in range(-3, 3)],
    'gamma': [10**x for x in range(-8, -4)],
}

rs = GridSearchCV(SVC(), params, n_jobs=-1, cv=10, verbose=2)
rs.fit(x_train, y_train)
print(rs.best_score_)
print(rs.best_params_)
```

```python
from sklearn.metrics import accuracy_score

svm = SVC(**rs.best_params_)
svm.fit(x_train, y_train)

print('Train score:', accuracy_score(y_train, svm.predict(x_train)))
print('Test score:', accuracy_score(y_test, svm.predict(x_test)))
```
