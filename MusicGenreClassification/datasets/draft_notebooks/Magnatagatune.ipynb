```python
import numpy as np
import pandas as pd
```

```python
csv_file = '/share/Magnatagatune/annotations_final.csv'
genres = '''classical, techno, electronic, rock, indian, opera, pop, classic, new age, dance, country, metal'''
genres = [x.strip() for x in genres.split(',')]
genres
```

```python
df = pd.read_csv(csv_file, sep='\t')
df
```

```python
synonyms = [
    ['beat', 'beats'], 
    ['chant', 'chanting'], 
    ['choir', 'choral'], 
    ['classical', 'clasical', 'classic'], 
    ['drum', 'drums'], 
    ['electro', 'electronic', 'electronica', 'electric'], 
    ['fast', 'fast beat', 'quick'], 
    ['female', 'female singer', 'female singing', 'female vocals', 'female voice', 'woman', 'woman singing', 'women'], 
    ['flute', 'flutes'], 
    ['guitar', 'guitars'], 
    ['hard', 'hard rock'], 
    ['harpsichord', 'harpsicord'], 
    ['heavy', 'heavy metal', 'metal'], 
    ['horn', 'horns'], 
    ['india', 'indian'], 
    ['jazz', 'jazzy'], 
    ['male', 'male singer', 'male vocal', 'male vocals', 'male voice', 'man', 'man singing', 'men'], 
    ['no beat', 'no drums'], 
    ['no singer', 'no singing', 'no vocal','no vocals', 'no voice', 'no voices', 'instrumental'], 
    ['opera', 'operatic'], 
    ['orchestra', 'orchestral'], 
    ['quiet', 'silence'], 
    ['singer', 'singing'], 
    ['space', 'spacey'], 
    ['string', 'strings'], 
    ['synth', 'synthesizer'], 
    ['violin', 'violins'], 
    ['vocal', 'vocals', 'voice', 'voices'],
    ['strange', 'weird']]
```

```python
df = df.iloc[:, 1:-1]
```

```python
data_df = df.copy()
for group in synonyms:
    group = sorted(group)
    intersection = set(group) & set(genres)
    if intersection:
        base = intersection.pop()
    else:
        base = group[0]
    group.remove(base)
    for synonym in group:
        data_df[base] += df[synonym]
print(df[genres].sum())
print(data_df[genres].sum())
```

```python
merged_data_df = data_df[genres]
merged_data_df
```
