```python
import os.path as osp
import os
import pandas as pd
```

```python
fpaths = ['/share/ISMIR2004/metadata/{}/tracklist.csv'.format(phase)
          for phase in ('training', 'development', 'evaluation')]
fpaths
```

```python
def read_csv(phase, path):
    df = pd.read_csv(path, header=None)[[0, 5]]
    df = df.rename(index=str, columns={0: 'genre', 5: 'path'})
    df['path'] = df['path'].map(lambda x: osp.join(phase, x))
    return df
dfs = [read_csv(phase, path) for phase, path in zip(('training', 'development', 'evaluation'), fpaths)]
```

```python
pd.concat(dfs, axis=0)
```
